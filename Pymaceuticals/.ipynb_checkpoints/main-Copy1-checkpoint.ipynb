```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_metadata = pd.read_csv(study_results_path)

# Combine the data into a single dataset
full_mouse_study = pd.merge(study_metadata, mouse_metadata, 
                                 how='outer', on='Mouse ID', sort=False)
# Display the data table for preview
full_mouse_study.sort_values(by="Timepoint", ascending=True).head()
```

```python
# Checking the number of mice.
print(full_mouse_study["Mouse ID"].nunique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dedup_mice = full_mouse_study.groupby(["Timepoint"])
dedup_mice["Mouse ID"].describe()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
full_mouse_study.loc[full_mouse_study["Mouse ID"] == "g989"]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
full_mouse_study_dedup = full_mouse_study[full_mouse_study["Mouse ID"] != 'g989']
mouse_study_df = full_mouse_study_dedup.drop_duplicates(["Mouse ID"]).reset_index(drop=True)

mouse_study_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
mouse_study_df["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, s.d., and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
mean_drug = full_mouse_study_dedup.groupby(["Drug Regimen"]).mean()
median_drug = full_mouse_study_dedup.groupby(["Drug Regimen"]).median()
variance_drug = full_mouse_study_dedup.groupby(["Drug Regimen"]).var()
std_drug = full_mouse_study_dedup.groupby(["Drug Regimen"]).std()
sem_drug = full_mouse_study_dedup.groupby(["Drug Regimen"]).sem()


# Assemble the resulting series into a single summary DataFrame.
drug_tumor_data = {"Mean Tumor Volume": mean_drug["Tumor Volume (mm3)"], "Median Tumor Volume": median_drug["Tumor Volume (mm3)"], 
                "Tumor Volume Variance": variance_drug["Tumor Volume (mm3)"], "Tumor Volume S.D.": std_drug["Tumor Volume (mm3)"], 
                "Tumor Volume SEM": sem_drug["Tumor Volume (mm3)"]}
drug_tumor_summary = pd.DataFrame(data=drug_tumor_data)
drug_tumor_summary
```

```python
# Using the aggregation method, produce the same summary statistics in a single line.
summary_table = full_mouse_study_dedup.groupby(["Drug Regimen"]).agg({"Tumor Volume (mm3)":["mean", "median", "var", "std", "sem"]})
summary_table
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
dedup_mice_finaltp = full_mouse_study_dedup.drop_duplicates(["Mouse ID"], keep='last', ignore_index=True)

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
mouse_tumorvol = pd.merge(mouse_study_df, dedup_mice_finaltp, 
                                 how='outer', on='Mouse ID', sort=False)
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Locate the rows which contain mice on each drug and get the tumor volumes

for treatment in treatments:
    final_tumor_vol = mouse_tumorvol.loc[mouse_tumorvol["Drug Regimen_x"] == treatment, "Tumor Volume (mm3)_y"]
    
    tumor_vol_data.append(final_tumor_vol)
    
    #Determine outliers
    quartiles = final_tumor_vol.quantile([.25,.5,.75])
    lowerq = quartiles[.25]
    upperq = quartiles[.75]
    iqr = upperq-lowerq

    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    
    # Determine outliers using upper and lower bounds
    outlier = final_tumor_vol.loc[(final_tumor_vol < lower_bound) | (final_tumor_vol > upper_bound)]
    print(f"{treatment} potential outlier: {outlier}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
Capomulin = []
Ramicane = []
Infubinol = []
Ceftamin = []

for row in mouse_tumorvol.index:
    if (mouse_tumorvol["Drug Regimen_x"][row]) == "Capomulin":
        Capomulin.append(mouse_tumorvol['Tumor Volume (mm3)_y'][row])
    elif (mouse_tumorvol["Drug Regimen_x"][row]) == "Ramicane":
        Ramicane.append(mouse_tumorvol['Tumor Volume (mm3)_y'][row])
    elif (mouse_tumorvol["Drug Regimen_x"][row]) == "Infubinol":
        Infubinol.append(mouse_tumorvol['Tumor Volume (mm3)_y'][row])
    elif (mouse_tumorvol["Drug Regimen_x"][row]) == "Ceftamin":
        Ceftamin.append(mouse_tumorvol['Tumor Volume (mm3)_y'][row])

data = [Capomulin, Ramicane, Infubinol, Ceftamin]
```

```python
fig,ax = plt.subplots()
plt.boxplot(data)
plt.xticks((1, 2, 3, 4), treatments)
plt.ylabel('Final Tumor Vol (mm3)')
plt.title('Final Tumor Volume by Treatment Group')
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
Capomulin_mouse_df = full_mouse_study.loc[full_mouse_study["Mouse ID"] == "i557"]

x_values = Capomulin_mouse_df['Timepoint']
y_values = Capomulin_mouse_df['Tumor Volume (mm3)']
plt.plot(x_values, y_values)
plt.xlabel('Time')
plt.ylabel('Tumor Vol (mm3)')
plt.title('Capomulin Treatment of Mouse i557')
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
Capomulin_mice = full_mouse_study.loc[full_mouse_study["Drug Regimen"] == "Capomulin"].groupby("Mouse ID").mean()

x_values = Capomulin_mice['Weight (g)']
y_values = Capomulin_mice['Tumor Volume (mm3)']
plt.scatter(x_values, y_values)
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Vol (mm3)')
plt.title('Capomulin: Average Tumor Volume vs. Weight')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

tumor_vol_mean = Capomulin_mice.iloc[:,1]
weight = Capomulin_mice.iloc[:,4]
corr = st.pearsonr(tumor_vol_mean,weight)

print(f'The correlation between mouse weight and the average tumor volume is {round(corr[0],2)}.')

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = 'y = ' + str(round(slope,2)) + 'x +' + str(round(intercept,2))
#y = 0.95x +21.55

plt.scatter(x_values, y_values)
plt.plot(x_values, regress_values, 'r-')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Vol (mm3)')
plt.title('Capomulin: Average Tumor Volume vs. Weight')
plt.annotate(line_eq, (30,30), color='red')
plt.show()
```
