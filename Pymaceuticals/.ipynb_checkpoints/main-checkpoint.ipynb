```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_metadata = pd.read_csv(study_results_path)

# Combine the data into a single dataset
full_mouse_study = pd.merge(study_metadata, mouse_metadata, 
                                 how='outer', on='Mouse ID', sort=False)
# Display the data table for preview
full_mouse_study.sort_values(by="Timepoint", ascending=True).head()
```

```python
# Checking the number of mice.
print(full_mouse_study["Mouse ID"].nunique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dedup_mice = full_mouse_study.groupby(["Timepoint"])
dedup_mice["Mouse ID"].describe()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
full_mouse_study.loc[full_mouse_study["Mouse ID"] == "g989"]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
full_mouse_study_dedup = full_mouse_study[full_mouse_study["Mouse ID"] != 'g989']
mouse_study_df = full_mouse_study_dedup.drop_duplicates(["Mouse ID"]).reset_index(drop=True)

mouse_study_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
mouse_study_df["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, s.d., and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
mean_drug = full_mouse_study_dedup.groupby(["Drug Regimen"]).mean()
median_drug = full_mouse_study_dedup.groupby(["Drug Regimen"]).median()
variance_drug = full_mouse_study_dedup.groupby(["Drug Regimen"]).var()
std_drug = full_mouse_study_dedup.groupby(["Drug Regimen"]).std()
sem_drug = full_mouse_study_dedup.groupby(["Drug Regimen"]).sem()


# Assemble the resulting series into a single summary DataFrame.
drug_tumor_data = {"Mean Tumor Volume": mean_drug["Tumor Volume (mm3)"], "Median Tumor Volume": median_drug["Tumor Volume (mm3)"], 
                "Tumor Volume Variance": variance_drug["Tumor Volume (mm3)"], "Tumor Volume S.D.": std_drug["Tumor Volume (mm3)"], 
                "Tumor Volume SEM": sem_drug["Tumor Volume (mm3)"]}
drug_tumor_summary = pd.DataFrame(data=drug_tumor_data)
drug_tumor_summary
```

```python
# Using the aggregation method, produce the same summary statistics in a single line.
summary_table = full_mouse_study_dedup.groupby(["Drug Regimen"]).agg({"Tumor Volume (mm3)":["mean", "median", "var", "std", "sem"]})
summary_table
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using Pandas.
mice_timepts = full_mouse_study_dedup.groupby(["Drug Regimen"]).sum()
mice_time_total = pd.DataFrame(data = mice_timepts["Timepoint"])
 
bar1 = mice_time_total.plot(kind='bar', facecolor='tab:blue', figsize=(5,4),
                           title="Timepoint Totals per Drug")
xticklabels = mice_time_total.index
bar1.set_xticklabels(xticklabels, rotation=90, rotation_mode='anchor', ha='right', wrap=True)


plt.savefig("output/Timepoints_by_Drug_bar_pandas.png")
plt.show()
```

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using pyplot.
x_axis = mice_timepts.index.tolist()
time_total = mice_timepts["Timepoint"]
tick_locations = [value for value in x_axis]

plt.bar(x_axis, time_total, color='tab:blue')

plt.xticks(rotation=90)
plt.title("Timepoint Totals per Drug")
plt.xlabel("Drug Regimen")

plt.savefig("output/Timepoints_by_Drug_bar_pyplot.png")
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using Pandas
mice_gender = mouse_study_df.groupby(["Sex"]).nunique()
mice_gender_total = pd.DataFrame(data = mice_gender["Mouse ID"])

pie1 = mice_gender_total.plot(kind='pie', autopct="%1.1f%%", labeldistance=1.1, subplots=True, figsize=(4,4),
                           title="Mouse Sex Distribution")

plt.savefig("output/Mice_Sex_Distribution_pie_pandas.png")
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
labels = ['Female', 'Male']
sizes = [123, 125]
colors = ['tab:blue', 'tab:orange']

plt.pie(sizes, labels=labels, colors=colors, autopct="%1.1f%%")

plt.title("Mouse Sex Distribution")

plt.savefig("output/Mice_Sex_Distribution_pie_pyplot.png")
plt.show()
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
dedup_mice_finaltp = full_mouse_study_dedup.drop_duplicates(["Mouse ID"], keep='last', ignore_index=True)

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
mouse_tumorvol = pd.merge(mouse_study_df, dedup_mice_finaltp, 
                                 how='outer', on='Mouse ID', sort=False)
mouse_tumorvol.head()
```

```python

mouse_tumorvol["Tumor Volume (mm3)_y"].describe()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
Capomulin_tumor_vol_data = []
Ramicane_tumor_vol_data = []
Infubinol_tumor_vol_data = []
Ceftamin_tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
quartiles = mouse_tumorvol['Tumor Volume (mm3)_y'].quantile([.25,.5,.75])
lowerq = quartiles[.25]
upperq = quartiles[.75]
iqr = upperq-lowerq

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)

print(quartiles)
print(f'IQR: {iqr}')

for row in mouse_tumorvol.index:
    if ((mouse_tumorvol["Tumor Volume (mm3)_y"][row]) > upper_bound) | ((mouse_tumorvol["Tumor Volume (mm3)_y"][row]) < lower_bound):
        print(f"Potential Outlier: {row}")
    
    
```

```python
# Locate the rows which contain mice on each drug and get the tumor volumes
for row in mouse_tumorvol.index:
    if (mouse_tumorvol["Drug Regimen_x"][row]) == "Capomulin":
        Capomulin_tumor_vol_data.append(mouse_tumorvol['Tumor Volume (mm3)_y'][row])
    elif (mouse_tumorvol["Drug Regimen_x"][row]) == "Ramicane":
        Ramicane_tumor_vol_data.append(mouse_tumorvol['Tumor Volume (mm3)_y'][row])
    elif (mouse_tumorvol["Drug Regimen_x"][row]) == "Infubinol":
        Infubinol_tumor_vol_data.append(mouse_tumorvol['Tumor Volume (mm3)_y'][row])
    elif (mouse_tumorvol["Drug Regimen_x"][row]) == "Ceftamin":
        Ceftamin_tumor_vol_data.append(mouse_tumorvol['Tumor Volume (mm3)_y'][row])
    
    # add subset 
Capomulin_tumor_vol_df = pd.DataFrame(Capomulin_tumor_vol_data, columns=["Tumor Volume (mm3)"])
Ramicane_tumor_vol_df = pd.DataFrame(Ramicane_tumor_vol_data, columns=["Tumor Volume (mm3)"])
Infubinol_tumor_vol_df = pd.DataFrame(Infubinol_tumor_vol_data, columns=["Tumor Volume (mm3)"])
Ceftamin_tumor_vol_df = pd.DataFrame(Ceftamin_tumor_vol_data, columns=["Tumor Volume (mm3)"])
    
    # Determine outliers using upper and lower bounds
#quartilesCa = Capomulin_tumor_vol_df.quantile([.25,.5,.75])
quartilesCA = st.iqr(Capomulin_tumor_vol_data)
print(quartilesCa)
lowerqCa = quartilesCa[0.25]
upperqCa = quartilesCa[0.75]
iqrCa = upperqCa-lowerqCa

lower_boundCa = lowerqCa - (1.5*iqr)
upper_boundCa = upperqCa + (1.5*iqr)

for value in Capomulin_tumor_vol_data:
    if (value > upper_bound) | (value < lower_bound):
        print(f"Capomulin's potential outliers: {value}")

quartilesR = Ramicane_tumor_vol_df.quantile([.25,.5,.75])    
lowerqR = quartilesR[.25]
upperqR = quartilesR[.75]
iqrR = upperqR-lowerqR

lower_boundR = lowerqR - (1.5*iqr)
upper_boundR = upperqR + (1.5*iqr)

for value in Ramicane_tumor_vol_data:
    if (value > upper_bound) | (value < lower_bound):
        print(f"Ramicane's potential outliers: {value}")

quartilesI = Infubinol_tumor_vol_df.quantile([.25,.5,.75])    
lowerqI = quartilesI[.25]
upperqI = quartilesI[.75]
iqrI = upperqI-lowerqI

lower_boundI = lowerqI - (1.5*iqr)
upper_boundI = upperqI + (1.5*iqr)

for value in Infubinol_tumor_vol_data:
    if (value > upper_bound) | (value < lower_bound):
        print(f"Infubinol's potential outliers: {value}")

quartilesC = Ceftamin_tumor_vol_df.quantile([.25,.5,.75])    
lowerqC = quartilesC[.25]
upperqC = quartilesC[.75]
iqrC = upperqC-lowerqC

lower_boundC = lowerqC - (1.5*iqr)
upper_boundC = upperqC + (1.5*iqr)

for value in Ceftamin_tumor_vol_data:
    if (value > upper_bound) | (value < lower_bound):
        print(f"Ceftamin's potential outliers: {value}")
```
